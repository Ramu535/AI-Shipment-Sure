## Support Vector Machine (SVM) Algorithm 
Support Vector Machine (SVM) is a supervised machine learning algorithm mainly used for:

- Classification (binary and multiclass)

- Regression (called SVR)

The core idea:
SVM tries to find the best decision boundary (called hyperplane) that separates different classes of data points with the maximum margin.
- Hyperplane

A line (in 2D), a plane (in 3D), or higher-dimensional surface that separates the data into classes.

SVM chooses the hyperplane that maximizes the separation (margin) between classes.

- Support Vectors

The data points closest to the hyperplane.

These are the critical elements of the dataset that influence the position and orientation of the hyperplane.

- Margin

The distance between the hyperplane and the nearest data points from each class.

SVM tries to maximize this margin for better generalization.

## Types of SVM

Linear SVM

- Used when data is linearly separable.

- Example: Classifying points with a straight line.

Non-linear SVM

- When data is not linearly separable, SVM uses the Kernel Trick to map data into higher dimensions where separation is possible.

###  What if data is not linearly separable?
When data is not linearly separable i.e it can't be divided by a straight line, SVM uses a technique called kernels to map the data into a higher-dimensional space where it becomes separable. This transformation helps SVM find a decision boundary even for non-linear data.

Common kernels:

- Linear Kernel

- Polynomial Kernel

- Radial Basis Function (RBF)

- Sigmoid Kernel
