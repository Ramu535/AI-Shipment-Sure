# SQL Queries using Pandas

### 1. Find the average cost of the product.

```python
df["Cost_of_the_Product"].mean()
```

### 2.Find the maximum discount offered.

```python
df["Discount_offered"].max()
```

### 3.Average rating by gender.

```python
df.groupby('Gender')['Customer_rating'].mean()
```

### 4.Average cost of product by warehouse

```python
df.groupby('Warehouse_block')['Cost_of_the_Product'].mean()
```

### 5. Average rating by gender.

```python
df.groupby('Gender')['Customer_rating'].mean()
```

### 6.number of prior purchases by gender.

```python
df.groupby("Gender")["Prior_purchases"].count()
```

### 7. heighest weight in every mode of shipment

```python
df.groupby("Mode_of_Shipment")["Weight_in_gms"].max()
```

### 8.Average customer rating per Warehouse_block and Mode_of_Shipment

```python
df.groupby(["Warehouse_block", "Mode_of_Shipment"])["Customer_rating"].mean()
```

```python
df["Reached.on.Time_Y.N"].value_counts()
```

 ### 9. Find top 5 heaviest products.

```python
df.nlargest(5,"Weight_in_gms").head(5)["ID"]
```

### 10.Find the top 2 mode of shipment that delivered the highest number of on-time deliveries.

```python
df[df['Reached.on.Time_Y.N'] == 1].groupby("Mode_of_Shipment")["Reached.on.Time_Y.N"].count().head(2)
```

### 11.For each warehouse, calculate the average number of care calls for late deliveries.

```python
df[df["Reached.on.Time_Y.N"]==0].groupby("Warehouse_block")["Customer_care_calls"].mean()
```

### 12.Find the percentage of high-importance products shipped by road.

```python
#d=df.groupby("Mode_of_Shipment")["Product_importance"].count().rename("count")
total_roads=df[(df["Mode_of_Shipment"]=="Road")].shape[0]
count=df[(df["Mode_of_Shipment"]=="Road") & (df["Product_importance"]=="high") ].shape[0]

print((count/total_roads)*100)
```

### 13.Identify the shipment mode with the lowest average weight


```python
a=df.groupby("Mode_of_Shipment")["Weight_in_gms"].mean()
min(a)
```

### 14.For each customer rating, find the average discount and average cost.

The .agg() method lets you apply aggregation functions (like mean, sum, count, etc.) to one or more columns.

```python
df.groupby("Customer_rating").agg({"Discount_offered":"mean","Cost_of_the_Product":"mean"})
```

### Count orders by product importance

```python
df.groupby("Product_importance").size()
## size() function when applied on groupby it will display the count of each group
```

df.size → total elements (rows × columns).

series.size → number of elements in Series.

df.groupby(...).size() → number of rows in each group.

```python
df['Product_importance'].value_counts()
# it is similar to above,but with standard function
```
