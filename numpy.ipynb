### Array Operations

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a + b)         
print(a * b) 
print(a-b)        
print(a ** 2)         
```

### Matrix Operations:

```python
np.dot(a,b)
```

```python
b = np.array([[4],[5],[2]])
np.matmul(a,b)
```

```python
print(np.sqrt(arr))
print(np.exp(arr))
print(np.log(arr))
print(np.sin(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
print(np.min(arr), np.max(arr))
```

### Linear Algebra

```python
mat = np.array([[1,2],[3,4]])
print(np.linalg.det(mat) )    # determinant
print(np.linalg.inv(mat)   )  # inverse
print(np.linalg.eig(mat)    ) # eigenvalues & eigenvectors
```

```python
print("Eigen values: ",np.linalg.eig(mat)[0]) 
print("Eigen vectors: ",np.linalg.eig(mat)[1]) 
```
